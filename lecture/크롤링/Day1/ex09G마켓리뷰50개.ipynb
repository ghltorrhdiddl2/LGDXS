```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
import time
```

```python
# 사이트 실행 및 요청
driver = uc.Chrome()
driver.get("https://item.gmarket.co.kr/Item?goodscode=2691771311")
```

```python
# 상품명 수집
title = driver.find_element(By.CSS_SELECTOR,"#itemcase_basic > div > div.box__item-info > h1")
title.text
```

```python
# 상품평 탭 클릭
driver.find_element(By.CSS_SELECTOR,"#container > div.vip-tabwrap.uxetabs > div.vip-tabnavi.uxeposfix > ul > li:nth-child(2) > a").click()
```

```python
# 리뷰데이터(내용) 5개 수집
review = driver.find_elements(By.CSS_SELECTOR,"#premium-wrapper > table > tbody > tr > td.comment-content > a > p.con")
for i in review:
    print(i.text)
```

```python
# 2번째 페이지 클릭
num = driver.find_elements(By.CSS_SELECTOR,"#premium-pagenation-wrap > div.board_pagenation > ul > li > a")
num[1].click()
```

```python
# 10번 반복하는 수집코드 통합
# 페이지가 클릭해서 변하는 구조에서는 반드시 클릭을 먼저하고 수집하자
# Why? 페이지가 넘기고 수집하지 않으면, 마지막 페이지가 수집이 되지 않는다
review_list = []
for i in range(10):
    num = driver.find_elements(By.CSS_SELECTOR,"#premium-pagenation-wrap > div.board_pagenation > ul > li > a")
    num[i].click()
    time.sleep(0.5)
    review = driver.find_elements(By.CSS_SELECTOR,"#premium-wrapper > table > tbody > tr > td.comment-content > a > p.con")
    for j in review:
        review_list.append(j.text)
```

- 페이지 클릭을 먼저하고 리뷰 수집을 하자

## time기능을 적절한 곳에 사용하는 방법
- 목적 : 요청과 응답 사이에 데이터를 받는 시간을 부여하기 위해서 time을 사용
- 올바른 사용법 : 클라이언트가 서버에게 요청을 한 다음에 반드시 time을 부여하자
- ex) 검색, 페이지 넘김 등과 같은 액션에 발생할 때 반드시 시간을 넣어주자

```python
review_list
```

```python
# 50페이지까지 수집
# 1페이지 클릭 -> 수집 -> 2페이지 클릭 -> 수집-----10페이지 수집-> 다음페이지 클릭(>)

review_list = []
for _ in range(5):
    for i in range(10):
        num = driver.find_elements(By.CSS_SELECTOR,"#premium-pagenation-wrap > div.board_pagenation > ul > li > a")
        num[i].click()
        time.sleep(0.5)
        review = driver.find_elements(By.CSS_SELECTOR,"#premium-wrapper > table > tbody > tr > td.comment-content > a > p.con")
        for j in review:
            review_list.append(j.text)
    driver.find_element(By.CSS_SELECTOR,"#premium-pagenation-wrap > div.board_pagenation > a.next").click()
    time.sleep(0.5)
```

## 이번 파일 정리
- 브라우저를 실행시 로봇으로 감지되는 경우에는 uc(driver = uc.Chrome())를 활용
- 페이지를 넘기는 구조는 크게 2가지 방식
- 1) 페이지의 넘버를 클릭해서 이동하는 방법
  2)  '>' 버튼을 통해서 이동하는 방법
- 반복문을 작성할 때는 코드를 먼저 완성하자 -> 반복문으로 감싸서 범위만 지정하자
- time을 적절하게 활용하자 -> 요청이 발생할 때 마다 작성/액션이 발생했을때 넣어주기
