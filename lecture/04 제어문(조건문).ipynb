#### 상황을 제어하는 문법
- 1. 조건문 : 상황에 따라 실행 흐름을 다르게 하는 문법
  2. 종류 :
     - if: 첫 번째 확인할 조건
     - elif: 첫 번째 조건이 아닐 경우 확인할 2번째 조건(조건 명확히 지정)
     - else: 위의 조건들이 모두 아닐 경우 확인하게 되는 마지막 관문(조건지정x)

```python
# if 키워드만 주어질 때
if True:
    print('지영썜과 함께하는 LG전자 DX스쿨')
```

```python
# if는 한 개
# 조건은 False
if False:
    print('지영썜과 함께하는 LG전자 DX스쿨')
```

```python
mask = int(input('마스크 개수 입력 >> '))

eight = 0; five = 0

eight = mask//8
rest = mask%8

if rest ==0:
    five = 0
elif rest <= 5: 
    five = 1 
else: five = 2

print(f'8개들이 포장지 개수 : {eight}개 \n5개들이 포장지 개수 : {five}개')
```

```python
score = int(input('점수 입력 >> '))

if score >= 90:
    print(f'{score}점은 A학점 입니다.')
elif score >= 80:
    print(f'{score}점은 B학점 입니다.')
elif score >= 70:
    print(f'{score}점은 C학점 입니다.')
elif score >= 60:
    print(f'{score}점은 D학점 입니다.')
else:
    print(f'{score}점은 F학점 입니다.')
```

```python
print('--------[SM Coffee]---------')
money = int(input('금액을 입력하세요 >> '))
coffee = int(input('커피를 고르세요.\n[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))

if coffee == 1:
    money -= 600
elif coffee == 2:
    money -= 1000
elif coffee == 3:
    money -= 1300
else:
    print('해당 커피가 없습니다.')

if money<0:
    print('돈이 부족해요.')
else:
    print(f'거스름돈: {money}원')
    print(f'천원: {money//1000}장')
    remain = money%1000
    print(f'오백원: {remain//500}개')
    remain %=500
    print(f'백원: {remain//100}개')
```
