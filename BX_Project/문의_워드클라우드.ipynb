```python
refri_cu_word=[];tv_cu_word=[];washingM_cu_word=[];dryer_cu_word=[];airC_cu_word=[]

for i in refri_cu:
    if any(word in i for word in contain_word):  # 관련 단어가 들어가 있으면 리스트에 추가
        refri_cu_word.append(i)
```

## 11번가와 쿠팡 QnA 합치기

```python
qna_refri = refri+refri_cu_word
qna_tv = tv+tv_cu_word
qna_washingM = washingM+washingM_cu_word
qna_dryer = dryer+dryer_cu_word
qna_airC = airC+airC_cu_word
```

### 토큰화

```python
qna_refri_token = [token for doc in qna_refri for token in word_tokenize(doc)]
qna_tv_token = [token for doc in qna_tv for token in word_tokenize(doc)]
qna_washingM_token = [token for doc in qna_washingM for token in word_tokenize(doc)]
qna_dryer_token = [token for doc in qna_dryer for token in word_tokenize(doc)]
qna_airC_token = [token for doc in qna_airC for token in word_tokenize(doc)]
```

## 데이터 클렌징

```python
p = re.compile("[!.,?;)(:ㅡㅋㅎㅇㅠㅜ\\^~a-zA-Z]+")
refri_clean = []

for doc in tqdm(qna_refri_token):
    if p.search(doc):
        continue
    refri_clean.append(doc)

refri_clean
```
