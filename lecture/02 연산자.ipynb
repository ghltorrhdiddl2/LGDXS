## 연산자 종류


### 산술연산자

```python
n1=10;n2=20
print(n1+n2)
print(n1-n2)
print(n1*n2)
print(n1/n2)
```

```python
n3=10;n4=7

print(n3/n4)
print(n3//n4)
print(n3%n4)
```

```python
# 문자열 데이터간 연산
str1 = '안녕'
str2 = '하세요.'

print(str1+str2)
print(str1+str2+str1)

# print(str1-str2)  ==> 에러발생
# print(str1*str2)  ==> 에러발생
# print(str1/str2)  ==> 에러발생
```

- 문자열 데이터간 산술연산은 -> 더하기 연산만 가능

```python
# 문자열과 숫자 산술연산

num1=10; str3='7'

# print(num1+str3)  ==> TypeError: unsupported operand type(s) for +: 'int' and 'str'
print(num1*str3)
# print(num1/str3)    ==> TypeError: unsupported operand type(s) for /: 'int' and 'str'
```

- 문자열과 숫자 산술연산 -> 곱하기 연산만 가능

```python
# 데이터 타입 변환

print(num1+int(str3))
print(str(num1)+str3)
```

```python
# input() 함수 사용하여 데이터 입력받기
num1 = input(); num2 = int(input())

print(type(num1))

num1 = int(num1)

print(type(num1))

print('더하기 결과 :',num1+num2)
print('빼기 결과 :',num1-num2)
print('곱하기 결과 :',num1*num2)
print('나누기 결과 :',num1/num2)
```

```python
num1 = int(input('정수를 입력하세요 >>'))
num2 = int(input('정수를 입력하세요 >>'))
             
print(f'더하기 결과 : {num1+num2}')
print(f'빼기 결과 : {num1-num2}')
print(f'곱하기 결과 : {num1*num2}')
print(f'나누기 결과 : {num1/num2}')
```

```python
time = int(input('시간 초 입력 >> '))
hour = int(time/60/60)
minute = int(time//60)-(60*hour)
second = time%60
print('{}시간 {}분 {}초'.format(hour, minute, second))
```

```python
# 정석답안
time  = int(input('시간 입력 >> '))

hour = time//3600
minute = time % 3600 // 60
second = time % 60

print(f'{hour}시간 {minute}분 {second}초')
```

## 논리 연산자




```python
# not 연산자
a=3; b=7

print(not True)
print(not False)
print(not a<7)
print(not a==b)
```

```python
# and 연산자

print(3>5 and 10==20)
print(3>5 and 10<20)
print(3<5 and 10<20)
```

```python
# or 연산자

print(3>5 or 10==20)
print(3>5 or 10<20)
print(3<5 or 10<20)
```

## 삼항 연산자


```python
result = '참' if True else '거짓'
print(result)

result = '참' if 3>5 else '거짓'
print(result)
```

```python
score = int(input('점수 입력'))
# 점수가 60이상이면 합격, 아닐경우 불합격 출력

'합격' if score>60 else '불합격'
```

```python
a = int(input('정수 입력 >> '))
b = int(input('정수 입력 >> '))

print(f'a: {a}') if a>b else print(f'b: {b}')
```

```python
# 홀수 짝수 판별하시오

num = int(input('정수 입력 >> '))
result = '홀수' if num%2 != 0 else '짝수'

print(f'{num}는(은) {result}입니다.')
```

```python
ball = int(input('농구공 수 입력 >> '))
basket = ball//5
result = basket+1 if ball%5 !=0 else basket

print(f'필요한 상자의 수 : {result}박스')
```

```python
worktime = int(input('근무시간 입력 >> '))

salary = 10000  # 시간당
add = salary*1.5 if worktime>8 else 0 #9시간부터 시간당

result = salary*worktime + add

print(f'총 지급액은 {int(result)}원 입니다.')
```
