```python
# 29
import pandas as pd
df = pd.read_csv("./part1/ch3/delivery_time.csv")
df
```

배달료 계산 기준표
- 5 미만 2000원
- 5~10미만 4000
- 10~15미만 6000
- 15이상 20미만 8000

```python
def delivery_fee(distance):
    if distance < 5:
        return 2000
    elif distance<10:
        return 4000
    elif distance<15:
        return 6000
    elif distance<20:
        return 8000

df['배달료'] = df['거리'].apply(delivery_fee)

df
```

```python
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['월'] = df['주문시간'].dt.month
df
```

```python
df.groupby('월')['배달료'].sum().sort_values()[::-1]
```
