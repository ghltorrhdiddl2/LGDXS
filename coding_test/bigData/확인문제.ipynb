```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "메뉴":['아메리카노','카페라떼','에스프레소','카페모카','바닐라라떼'],
    "가격":[4500,5000,4000,5900,5300],
    "칼로리":[10,110,np.NaN,210,np.NaN],
    "원두":['과테말라','브라질','과테말라',np.NaN, np.NaN]
})

df
```

```python
# Q1
df['칼로리'].fillna(df['칼로리'].min(), inplace=True)
df
```

```python
# Q2
df['원두'].fillna(df['원두'].mode()[0], inplace=True)
df
```
