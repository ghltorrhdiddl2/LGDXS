```python
from kiwipiepy import Kiwi
import re

pattern = r'[^0-9가-힣\s\.\?\!]'


kiwi = Kiwi()
new_reviews = []
kiwi.space(df["review"])

# 특수문자 제거
for review in df["review"]:
  text = re.sub(pattern, "", review)
  new_reviews.append(text)

df["review_clean"] = new_reviews
```

```python
stopwords_df = pd.read_csv("./data/ko-stopwords.csv")
```

```python
# list 형변환 -> 불용어 제거에 활용!
stopwords_list = list(stopwords_df["stopwords"])
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt() # 객체 초기화 생성
```

```python
def pos_tagging(text):
  pos_words = okt.pos(text, stem = True,norm=True)

  tagged_list = [] # 결과 저장 리스트

  for word, tag in pos_words:
    if tag in ["Noun","Adjective", "Verb"]:   # 명사, 동사, 형용사 인가?
      if word not in stopwords_list:
        tagged_list.append(word)
  return tagged_list
```

```python
from tqdm.auto import tqdm
```

```python
# 각 리뷰에 접근해서 정제 된 키워드들만 추출
tagged_review = []
for review in tqdm(df["review_clean"]):
  tagged = pos_tagging(review)
  tagged_review.append(tagged)
```

```python
df["tagged_review"] = tagged_review
print(tagged_review[0])
```

```python
import gensim # 자연어 처리, 토픽 모델링을 위한 라이브러리
from gensim.models.doc2vec import TaggedDocument # 문서 태깅 클래스
```

```python
import ast
# 문서 태깅 수행
tagged_corpus = []
for i,tokens in enumerate(df["tagged_review"]):
  if isinstance(tokens, str):  # tokens가 문자열이면 변환
        tokens = ast.literal_eval(tokens)  # 문자열을 리스트로 변환
  tag = f'document {i}'
  Tagged_doc = TaggedDocument(tags = [tag],words = tokens)
  tagged_corpus.append(Tagged_doc)
```

```python
from gensim.models import doc2vec
```

```python
# 1. 모델 초기화
model = doc2vec.Doc2Vec(vector_size = 150,  # 만들고자 하는 벡터의 차원 수
                        alpha = 0.025,      # 초기 학습률
                        min_alpha = 0.001,  # 학습 진행중에 점차 감소시키는 최소 학습률
                        window = 8,         # 주변 단어 범위 설정(앞뒤 8개 단어)
                        min_count = 5 )
```

```python
# 2. 단어사전 생성
model.build_vocab(tagged_corpus)
```

```python
print(len(model.wv))  # 남아 있는 단어 개수 확인
```

```python
model.corpus_count # 모델이 인식하고 있는 문서의 개수
```

```python
# 3. 학습
model.train(tagged_corpus,
            total_examples=model.corpus_count, # 총 문서 개수
            epochs= 5)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# dendrogram : 계층적 군집 결과를 트리 형태로 시각화
# linkage : 계층적 군집 수행 -> 군집 방식(ward)
```

```python
# 계층적 군집 수행
linked = linkage(list(df["vector"]), # 군집화 할 데이터
        "ward")                # 군집 방식 -> ward : 군집 내 분산이 최소화
```

```python
linked.shape
# 1018 -> 1019개의 문서가 있음 -> 계층적 군집 수행 시 1018번의 병합이 일어났다!
```

```python
# 덴드로그램 시각화
plt.figure(figsize=(15,10))

dendrogram(linked,                     # 계층적 군집 결과
           orientation= "top",         # 그리는 방향(위에서 아래로)
           distance_sort="descending", # 클러스터 간 거리를 기준으로 내림차순으로 정렬
           show_leaf_counts=True,      # 말단 노드마다 샘플 개수 출력
           color_threshold=7)         # y축 기준 -> 18 이상은 색상으로 표현하지 않겠다!


plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering # 병합 군집
from sklearn.metrics.cluster import silhouette_score # 평균 실루엣 계수 계산하는 도구
```

```python
n_clusters = range(2,7)
scores = []

for n in n_clusters:
  cluster = AgglomerativeClustering(n_clusters= n, linkage = "ward")
  cluster_label = cluster.fit_predict(list(df["vector"]))
  score = silhouette_score(list(df["vector"]), cluster_label)
  scores.append(score)
```

```python
all_doc = []
for i in df["cluster"].unique():
  # 각 클러스터를 하나의 문서로 가정
  # df에서 각 문서에 대해 불리언 인덱싱 수행 -> 키워드 접근
  pos_tagged = df[df["cluster"] == i]["tagged_review"]

  cluster_doc ="" # 동일 문서 내에 모든 키워드를 하나로 합치기 위한 비어있는 문자열 초기화
  for pos in pos_tagged:
    doc = " ".join(pos) # list에서 키워드들을 하나의 문자열로 형변환
    cluster_doc += doc  # 클러스터별 문서를 만들기 위해서 하나로 이어붙이기
  all_doc.append(cluster_doc)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# 초기화
vectorizer = TfidfVectorizer()

```

```python
# 단어사전 등록

tfidf_vector = vectorizer.fit_transform(all_doc)
```

```python
vectorizer.vocabulary_
```

```python
# 학습 된 단어사전 확인
feature_names = vectorizer.get_feature_names_out()
# array 형변환
tfidf_array = tfidf_vector.toarray()
# df로 형변환
tfidf_df = pd.DataFrame(tfidf_array,columns=feature_names).T
```

```python
# TFIDF 값이 높은 순으로 정렬 + 저장
for column in tfidf_df.columns:
  # 각 클러스터 별 TFIDF Series를 내림차순 정렬
  clustering_tfidf = tfidf_df[column].sort_values(ascending=False)
  # 컬럼으로 따로 인덱스 (단어사전들)들을 빼둔뒤, Series를 DF로 변환
  clustering_tfidf = clustering_tfidf.reset_index()
  clustering_tfidf.columns = ["word","tfidf"]
  # 저장
  clustering_tfidf.to_csv(f"./data/new/cluster_{column}_TFIDF.csv",
                          encoding = "UTF-8",index=False)
```
