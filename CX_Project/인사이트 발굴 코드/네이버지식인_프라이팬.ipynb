```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
driver = uc.Chrome()
driver.get("https://search.naver.com/search.naver?ssc=tab.kin.kqna&where=kin&sm=tab_jum&query=%ED%94%84%EB%9D%BC%EC%9D%B4%ED%8C%AC")
driver.maximize_window()
time.sleep(0.5)

# 스크롤링
body = driver.find_element(By.CSS_SELECTOR,"body")
n=0
while n<100:
    # 스크롤 내리기 전 데이터 저장
    first_data = driver.page_source
    body.send_keys(Keys.END)
    time.sleep(0.5)
    # 내린 후 데이터 저장
    last_data = driver.page_source
    # 멈추는 조건 부여
    if first_data == last_data:
        break
    n+=1
time.sleep(3)

pan_url = driver.find_elements(By.CSS_SELECTOR,"div.api_subject_bx > ul > li > div > div.question_area > div:nth-child(3) > a")
pan_url_list = []; pan_review_list = []
for i in pan_url:
    pan_url_list.append(i.get_attribute('href'))

len(pan_url_list)
```

```python
pan_url_list
```

```python
pan_url_list[0]
```

```python
pan_list = []

for i in range(len(pan_url_list)):
    driver.get(pan_url_list[i])
    title = driver.find_element(By.CSS_SELECTOR,"#content > div.endContentLeft._endContentLeft > div.contentArea._contentWrap > div.endTitleSection")
    content = driver.find_element(By.CSS_SELECTOR,"#content > div.endContentLeft._endContentLeft > div.contentArea._contentWrap > div.questionDetail")
    pan_list.append(title.text[3:]+content.text)

driver.quit()
```

```python
pan_list[:10]
```

```python
df = pd.DataFrame({'content':pan_list})
df.head()
```

```python
df.info()
```

```python
df.to_csv("pan_naver.csv", index=False,encoding='utf-8-sig')
```
