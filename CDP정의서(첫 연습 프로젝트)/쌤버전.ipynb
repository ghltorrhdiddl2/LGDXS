```python
driver = wb.Chrome()
driver.get("https://prod.danawa.com/info/?pcode=3521853&keyword=lg+%EA%B3%B5%EA%B8%B0%EC%B2%AD%EC%A0%95%EA%B8%B0&cate=1022486#bookmark_cm_opinion")
action = ActionChains(driver)
time.sleep(1)
re_list = []
for i in range(10) :
    body = driver.find_element(By.CSS_SELECTOR,"body")
    body.send_keys(Keys.END)
    time.sleep(0.5)
    re = driver.find_elements(By.CSS_SELECTOR,'div > div.atc')
    for j in re :
        re_list.append(j.text)
    page_num = driver.find_elements(By.CSS_SELECTOR,"#danawa-prodBlog-companyReview-content-list > div > div > div > a")
    action.move_to_element(page_num[0]).perform()
    time.sleep(0.5)
    #page_num[i].click()
    time.sleep(1)
    driver.quit()
re_list
```
