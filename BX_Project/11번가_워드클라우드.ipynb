### 데이터 클렌징

```python
import re
p = re.compile("[!.,?;)(:ㅡㅋㅎㅇㅠㅜ\\^~a-zA-Z]+")
refri_clean = []

for doc in tqdm(refri_token):
    if p.search(doc):
        continue
    refri_clean.append(doc)

refri_clean
```

### 형태소 분석

```python
from kiwipiepy import Kiwi
Kiwi = Kiwi()
```

```python
refri_clean_morphs = [] # 전체리스트 생성

for doc in tqdm(refri_clean):
    doc_merge = ''.join(doc)
    doc_merphs = Kiwi.tokenize(doc_merge)  
    morphs_list = []
    for t in doc_merphs:
        if t.tag in ["NNG","VV","VA"]: # 일반명사, 동사, 형용사만 추출
            morphs_list.append(t.form)
    refri_clean_morphs.append(' '.join(morphs_list))
```

```python
refri_clean_morphs
```

```python
token_list_morphs = [t for doc in refri_clean_morphs for t in word_tokenize(doc)]
counter_morphs = Counter(token_list_morphs)
counter_morphs.most_common(100)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(background_color='white', random_state=20, font_path='C://Windows/Fonts/LG_Smart_UI-SemiBold.ttf')
freq_dict = {k:v for k,v in counter_morphs.most_common(50)}
freq_dict
```
