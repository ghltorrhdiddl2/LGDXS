## HR 대시보드 샘플데이터 구성하기

```python
import random
 
# 가상의 인사 데이터 생성 함수 (딕셔너리 안에 딕셔너리 형태)
def generate_employee_dict_data(num_records=10):
    genders = ["남", "여"]
    marital_statuses = ["미혼", "기혼"]
    positions = ["사원", "대리", "과장", "차장", "부장"]
    majors = ["경영학", "컴퓨터공학", "디자인", "심리학", "화학공학", "기계공학", "국문학", "영문학", "전자공학", "의학"]
    work_types = ["재택", "출퇴근"]
    work_part = ['기획', '총무', '홍보마케팅', '비서실', '운영지원']
    employment_statuses = ["재직", "퇴사", '휴직']
    education_levels = ["고등학교 졸업","대학교 졸업", "대학원 졸업"]
   
    employees = {}
    for _ in range(num_records):
        employee_id = random.randint(1000, 9999)
        work_type = random.choice(work_types)
        commute_distance = random.randint(1, 50) if work_type == "출퇴근" else None
        education = random.choice(education_levels)
        major_choice = random.choice(majors) if education in ["대학교 졸업", "대학원 졸업"] else None
 
        employees[employee_id] = {
            "성별": random.choice(genders),
            "나이": random.randint(25, 60),
            "혼인여부": random.choice(marital_statuses),
            "부서" : random.choice(work_part),
            "직급": random.choice(positions),
            "급여(월)": random.randint(3000, 5000) * 1000,
            "학력": education,
            "전공": major_choice,
            "근속 연차": random.randint(1, 30),
            "근무 환경 만족도": random.randint(1, 5),
            "근무형태": work_type,
            "출퇴근 거리": commute_distance,
            "잔여휴가 갯수": random.randint(0, 25),
            "초과근무 시간(월 총계)" : random.randint(0, 30),
            "재직여부": random.choice(employment_statuses),
           
        }
    return employees
 
# 데이터 생성
employee_dict_data = generate_employee_dict_data(10)
 
# 결과 출력
import pprint
pprint.pprint(employee_dict_data)
```

### 데이터를 파일로 저장하기: pickle 모듈

```python
import pickle

profile_file = open('profile.pickle', 'wb')

profile = {'이름':'스누피', '나이':30, '취미':['축구','골프','코딩']}
print(profile)
```

- 작성한 데이터를 dump(저장할 데이터, 저장할 파일명) 함수를 사용해 파일로 저장
- 마지막에 close() 함수로 파일 닫기

```python
pickle.dump(profile, profile_file)  # profile 데이터를 profile_file 파일에 저장
profile_file.close() # 파일 닫기
```

- 데이터가 잘 저장 되었는지 확인
> load(불러올 파일명)

```python
profile_file = open('profile.pickle', 'rb')  # r: 읽어오기
profile = pickle.load(profile_file)

print(profile)
profile_file.close()
```
