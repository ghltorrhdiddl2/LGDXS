### 데이터 정리


```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[0,0,1,1,1,0,1,0], 'B':[0,1,0,1,0,0,1,1,], 'A and B':[0,0,0,1,0,0,1,0]})
df
```

- 데이터 전처리
    - 문제(X), 답(y) 분리하기
    - 문제 : 0 0 / 1 0 / 0 1 / 1 1
    - 답 : 0 / 0 / 0 / 1
<br><br>
  - AND 규칙을 내부적으로 정립! 패턴을 파악했다!
  - 잘 패턴을 파악했는지 검증(test) : 예측진행하자!
  - 학습용, 테스트용 분리하기
  - 훈련문제, 테스트 문제, 훈련답, 테스트답 : X_train, X_test, y_train, y_test
  - 데이터의 크기를 확인해야 함(shape)

```python
X = df[['A','B']]
y = df['A and B']

# 훈련용, 테스트용
X_train = X[:6]
y_train = y[:6]

X_test = X[6:]
y_test = y[6:]
```

```python
print('훈련용', X_train.shape, y_train.shape)
print('테스트용', X_test.shape, y_test.shape)
```

### 머신러닝 모델 학습하기

```python
# sklearn : 과학적 도구의 모음
# science kit learn 모듈
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 모델생성
# n_neighbors=1 매개변수를 1로 조정해보자
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
# 모델이 학습할 수 있을까?
# data(X_train,y_train) -> fit -> modle
knn_model.fit(X_train,y_train)
```

```python
# 새로운 테스트 문제가 주어질때 결과를 뽑아낼까? 예측할 수 있나?
# model <- data(테스트문제)
# predict()
pred = knn_model.predict(X_test)
```

```python
# 2개의 문제에 대해서 2개의 예측값을 출력함!
# 결과값이 여러개일 때 잘 맞췄는지 평가해주는 도구를 활용
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred) # 백분율 1.0*100(%)
```
