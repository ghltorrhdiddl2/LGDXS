## !!와인 추천!!

```python
sweet = int(input('당도 설정 (1-5)>> '))
body = int(input('바디감 설정 (1-5)>> '))
sparkl = int(input('탄산 포함? 미포함? (1,0)>> '))

condition = (df['Sweetness']==sweet)&(df['Body']==body)&(df['Sparkling']==sparkl)
result = df.loc[condition, ['Sweetness','Acidity','Body','Tannin','Sparkling','Price','Grade']]
result
# print('추천 와인 명 >> ', )
```

```python
print('현재 등록된 향은 Vanilla, Fruit, Floral, Woody, Herbal 이 있습니다.')

f= input('선호하는 향을 입력해주세요 >> ')

flavor = df['Flavor Profile'].str.contains(f)
flavor_df = df.loc[flavor,['Flavor Profile','Price','Grade']]

print(flavor_df.shape)
flavor_df
```

```python
print('현재 등록된 와인 종류는 White, Dessert, Red, Sparkling, Rose 가 있습니다.')
t = input('선호하는 와인 종류를 입력해주세요 >> ')
w_type = df['Type'].str.contains(t)
w_type_df = df.loc[w_type, ['Type','Sweetness','Acidity','Body','Tannin','Sparkling','Price','Grade']]

w_type_df
```

```python
print('선택하신 와인 타입과 향 추천 리스트')

condition = w_type & flavor
type_flavor = df.loc[condition, ['Type','Flavor Profile','Alcohol Content','Price','Grade']]
type_flavor
```

## 만약 추천한다면 경로

```python
ff = input('곁들여 즐길 와인 추천 (Meat, Dessert, Seafood) >> ')

food = df['Food Pairing'].str.contains(ff)
food_df = df[food]

print(food_df.shape)
food_df
```

```python
# 선택지 1

sweet = int(input('당도 설정 (1-5)>> '))
body = int(input('바디감 설정 (1-5)>> '))
sparkl = int(input('탄산 포함? 미포함? (1,0)>> '))

condition = (food_df['Sweetness']==sweet)&(food_df['Body']==body)&(food_df['Sparkling']==sparkl)
result = food_df[condition]
result
```

```python
recommend = result.index.to_list()

print('추천 와인 ID 및 이름')
for i in recommend:
    print(i,result.loc[i]['Name'])
```

```python
# 선택지 2

print('현재 등록된 향은 Vanilla, Fruit, Floral, Woody, Herbal 이 있습니다.')

f= input('선호하는 향을 입력해주세요 >> ')

flavor = food_df['Flavor Profile'].str.contains(f)
flavor_df = food_df[flavor]

print(flavor_df.shape)
flavor_df
```

```python
print('현재 등록된 와인 종류는 White, Dessert, Red, Sparkling, Rose 가 있습니다.')
t = input('선호하는 와인 종류를 입력해주세요 >> ')
w_type = flavor_df['Type'].str.contains(t)
w_type_df = flavor_df[w_type]

w_type_df
```

```python
recommend2 = w_type_df.index.to_list()

print('추천 와인 ID 및 이름')
for i in recommend2:
    print(i,w_type_df.loc[i]['Name'])
```
