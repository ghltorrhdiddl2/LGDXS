- 함수를 사용하는 이유
  + 반복적인 프로그래밍을 피할 수 있다.

  + 프로그램에 문제가 발생하거나 기능의 변경이 필요할 때에도 손쉽게 유지보수가 가능하다.

```python
# 두개의 숫자를 매개변수로 받아 더하기 결과값을 반환 하는 함수
def number_sum(num1, num2):
    result = num1 + num2
    return result
```

```python
a=number_sum(1,2)
print(a)
print(number_sum(15,20))
```

```python
def cal(num1, num2, op):
    if op == '+':
        return num1+num2
    elif op == '-':
        return num1-num2

num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))
op = input('연산자 입력(+,-) >> ')

result = cal(num1, num2, op)
print('결과 : {}'.format(result))
```

```python
def divisor(n):
    arr = []
    for i in range(1, n+1):
        if n % i == 0:
           arr.append(i)
    return arr
```

```python
print(divisor(36))
```
