### 인덱싱과 슬라이싱
- 파이썬의 문자열은 각 글자의 위치를 나타내는 인덱스가 부여된다
- 문자열의 왼쪽부터 부여되는 인덱스는 0,1,2,3,4,5 등으로 부여되고
- 문자열의 오른쪽부터 부여되는 인덱스는 -1,-2,-3,-4,= 등으로 부여된다
- 특정 문자 1개를 접근하는 방법을 인덱싱, 여러 문자를 접근하는 방법을 슬라이싱이라고 한다

```python
# 인덱싱
s = 'My name is SJ'

# 3번째 인덱스에 있는 n 추출하기
print(s[3])
# 음수 인덱스로 접근하기
print(s[-10])
```

```python
# 슬라이싱 -> [시작인덱스:끝인덱스]
# 끝인덱스는 +1 값만큼 증가시켜서 써주기
# name 글자 추출하기

s = 'My name is SJ'

s[3:7]
```

```python
# 시작인덱스와 끝인덱스는 생략 가능
# 시작인덱스 생략 -> 맨 처음부터
# 끝인덱스 생략 -> 맨 끝까지

print(s[:5])
print(s[7:])

# 시작과 끝을 모두 생략 -> 처음부터 끝까지 가져오기
print(s[:])
```

```python
day = '2020년 3월 3일의 날씨는 맑음입니다.'

print('날짜: ' + day[:11] + '\n날씨 : ' + day[-6:-4])
```

#### format 함수를 사용한 포매팅

```python
month =7
day =7
s='오늘은 {}월{}일 입니다.'.format(month,day)
s
```
