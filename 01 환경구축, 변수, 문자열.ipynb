### 변수명 규칙
1. 영문자, 숫자, 언더바(_)를 사용할 수 있다. (단, 대소문자 구분)
2. 숫자로 시작되는 변수명은 사용이 불가능한다.
3. 키워드 사용은 불가능하다. (키워드 : 내장함수를 변수명으로는 x)

### 변수명 규칙(권장사항)
- 변수명의 첫 글자는 항상 소문자로 만든다.
- 두가지의 문자를 섞어서 변수명을 만들 경우 두 단어를 구분할 수 있게 한다. (nameNum, name_naum)

### 변수 대입 방법

```python
# 변수 a에 10을 대입, 변수 b에 15를 대입
a=10 
b=15
print(a,b)

a,b = 12,17
print(a,b)
```

```python
# 변수 str1과 str2에 'python' 문자열을 대입하시오
str1, str2 = 'python','python'
print(str1,str2)

str1 = str2 = 'Python'
print(str1,str2)
```

## 문자열

```python
# 문자열 사용시 주읭할 것 : 인용부호 사용
# 문자열 안에 인용부호를 포함시키고 싶을 때 -> 이스케이프 코드 활용 \(역슬래쉬)

s1 = 'she\'s gone'
print(s1)

s2 = 'she\"s gone'
print(s2)
```

```python
# 인용부호 포함 방법2
s1 = "she's gone"
print(s1)

s2 = 'she"s gone'
print(s2)
```

```python
# 개행 포함
s3= "안녕하세요. \n저는 심수지 입니다."
s4= '''안녕하세요.
저는 심수지 입니다.'''
print(s3)
print(s4)
```
